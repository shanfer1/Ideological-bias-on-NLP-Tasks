```python
def make_queries(username, words):
    query = 'from:'+username+ " "
    res=''
    for y in words:
        res=res+'"'+y+'"'+ ' OR '
    res=res[:len(res)-4]
    query += res
    return "'"+query+"'"
```

```python
def collect_data(username):
    api = get_api()
    # date_since = ''
    db = pd.DataFrame(columns=['screenName',
                               'description',
                               'location',
                               'following',
                               'followers',
                               'totaltweets',
                               'retweetcount',
                               'text',
                               'topic'
                               ], dtype=object)
    for topic, words in topic_hashtags.items():
        res = make_queries(username, words)
        print(res)
        tweets = tweepy.Cursor(api.search_tweets,
                                   res, lang="en",
                                   tweet_mode='extended').items(200)
        i = 1
        for tweet in tweets:
            username = tweet.user.screen_name
            description = tweet.user.description
            location = tweet.user.location
            following = tweet.user.friends_count
            followers = tweet.user.followers_count
            totaltweets = tweet.user.statuses_count
            retweetcount = tweet.retweet_count
            tweet_text = tweet.full_text
            topic_classified = topic
            ith_tweet = [username, description,
                     location, following,
                     followers, totaltweets,
                     retweetcount, tweet_text, topic_classified]
            db.loc[len(db)] = ith_tweet
            i = i + 1
    return db
```

```python
#label encoding the categories. After this each category would be mapped to an integer.
def encode_topic(dataframe):
    encoder = LabelEncoder()
    dataframe['topicEncoded'] = encoder.fit_transform(dataframe['topic'])
    dataframe['text'] = dataframe['text'].apply(lambda text: str(text).lower())
    dataframe['text_len'] = dataframe['text'].apply(lambda x: len(str(x).split()))
    return dataframe
```

```python
def regular_encode(texts, tokenizer, maxlen=512):
    enc_di = tokenizer.batch_encode_plus(
        texts, 
        return_attention_masks=False, 
        return_token_type_ids=False,
        pad_to_max_length=True,
        max_length=maxlen
    )
    
    return np.array(enc_di['input_ids'])
```

```python
X_train,X_test ,y_train,y_test = df2['text'], df3['text'], df2['topicEncoded'], df3['topicEncoded']
```

```python
#tokenizing the news descriptions and converting the categories into one hot vectors using tf.keras.utils.to_categorical
Xtrain_encoded = regular_encode(X_train.astype('str'), tokenizer, maxlen=80)
ytrain_encoded = tf.keras.utils.to_categorical(y_train, num_classes=5,dtype = 'int32')
Xtest_encoded = regular_encode(X_test.astype('str'), tokenizer, maxlen=80)
ytest_encoded = tf.keras.utils.to_categorical(y_test, num_classes=5,dtype = 'int32')
```

```python
def build_model(transformer, loss='categorical_crossentropy', max_len=512):
    input_word_ids = tf.keras.layers.Input(shape=(max_len,), dtype=tf.int32, name="input_word_ids")
    sequence_output = transformer(input_word_ids)[0]
    cls_token = sequence_output[:, 0, :]
    #adding dropout layer
    x = tf.keras.layers.Dropout(0.3)(cls_token)
    #using a dense layer of 5 neurons as the number of unique categories is 5. 
    out = tf.keras.layers.Dense(5, activation='softmax')(x)
    model = tf.keras.Model(inputs=input_word_ids, outputs=out)
    #using categorical crossentropy as the loss as it is a multi-class classification problem
    model.compile(tf.keras.optimizers.Adam(lr=3e-5), loss=loss, metrics=['accuracy'])
    return model
```

```python
#building the model on tpu
with strategy.scope():
    transformer_layer = transformers.TFAutoModel.from_pretrained('bert-large-uncased')
    model = build_model(transformer_layer, max_len=80)
model.summary()
```

```python
#creating the training and testing dataset.
BATCH_SIZE = 32*strategy.num_replicas_in_sync
AUTO = tf.data.experimental.AUTOTUNE 
train_dataset = (
    tf.data.Dataset
    .from_tensor_slices((Xtrain_encoded, ytrain_encoded))
    .repeat()
    .shuffle(2048)
    .batch(BATCH_SIZE)
    .prefetch(AUTO)
)
test_dataset = (
    tf.data.Dataset
    .from_tensor_slices(Xtest_encoded)
    .batch(BATCH_SIZE)
)
```

```python
#training for 10 epochs
n_steps = Xtrain_encoded.shape[0] // BATCH_SIZE
train_history = model.fit(
    train_dataset,
    steps_per_epoch=n_steps,
    epochs=10
)
```

```python
#making predictions
preds = model.predict(test_dataset,verbose = 1)
#converting the one hot vector output to a linear numpy array.
pred_classes = np.argmax(preds, axis = 1)
```
